```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

home_url = "https://kathmandupost.com"

def getHTMLdocument(url):
      
    response = requests.get(url)
      
    return response.text

def getData(url):
    html_document = getHTMLdocument(url)
    soup = BeautifulSoup(html_document, 'html.parser')

    # Title
    title = soup.title.text

    # Crawl content
    section = soup.find_all("section", class_="story-section")[0]
    content = section.text

    # Find date
    div = soup.find_all("div", class_="updated-time")[1]
    date = div.text.replace('Updated at : ', '')

    # Find category
    category = soup.find_all("h4", class_="title--line__red")[0].text

    return title, content, date, category
```

```python
def crawlKathmanduPOST():
    html_document = getHTMLdocument(home_url)

    soup = BeautifulSoup(html_document, 'html.parser')

    articles = soup.find_all("article")

    titles = []
    contents = []
    urls = []
    dates = []
    categories = []

    for article in articles:
        link = home_url + article.a['href']

        title, content, date, category = getData(link)

        if content == ' ':
            continue

        titles.append(title)
        contents.append(content)
        urls.append(link)
        dates.append(date)
        categories.append(category)

    
    data = {'title': titles, 'content':contents, 'date':dates, 'category': categories, 'url':urls}

    return data
```

```python
data = crawlKathmanduPOST()
```

```python
df = pd.DataFrame.from_dict(data)
df
```
